```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('GDPlist.csv', encoding = "ISO-8859-1")
```

```python
df.head()
```

```python
df.tail()
```

```python
# Lọc các quốc gia ở South America
df_SA = df[df['Continent'] == 'South America']
```

```python
# Sắp xếp dữ liệu theo GDP giảm dần
df_SA = df_SA.sort_values(by='GDP (millions of US$)', ascending=False)
```

```python
# Vẽ biểu đồ
plt.bar(df_SA['Country'], df_SA['GDP (millions of US$)'])
plt.xlabel('Country')
plt.ylabel('GDP (millions of US$)')
plt.title('So sánh GDP của các quốc gia ở Nam Mỹ')

# Đặt góc quay của tên quốc gia để dễ đọc
plt.xticks(rotation=45, ha='right')

# Hiển thị biểu đồ
plt.show()
```

Nhận xét
- Brazil là nước có GDP cao nhất và cao hơn hẳn các quốc gia khác ở Nam Mỹ
- Ecuador, Uruguay và Paraguay có GDP thấp nhất ở Nam Mỹ

```python
# Lọc dữ liệu cho các quốc gia Đông Nam Á
countries = ['Vietnam', 'Indonesia', 'Cambodia', 'Thailand', 'Malaysia']
sea_df = df[df['Country'].isin(countries)]
print(sea_df)
```

```python
# Tính tổng GDP của 5 nước Đông Nam Á
seasia_gdp = sea_df['GDP (millions of US$)'].sum()
```

```python
# Lọc dữ liệu cho Việt Nam
vietnam_df = sea_df[sea_df['Country'] == 'Vietnam']

# Tính tỷ lệ đóng góp của Việt Nam
contribution_ratio = (vietnam_df['GDP (millions of US$)'] / seasia_gdp) * 100

print(f'Tổng GDP của 5 nước Đông Nam Á: {seasia_gdp} millions of US$')
print(f'Tỷ lệ đóng góp của Việt Nam: {contribution_ratio}%')
```

```python
# Vẽ biểu đồ
plt.bar(sea_df['Country'], sea_df['GDP (millions of US$)'])
plt.xlabel('Country')
plt.ylabel('GDP (millions of US$)')
plt.title('So sánh GDP của các quốc gia ở Đông Nam Á')
```
