Có nhiều giá trị ngoại lai ở cột MonthlyIncome và DebtRatio

```python
# Loại bỏ giá trị ngoại lai (cột MonthlyIncome)
# tính giá trị Q1 và Q3
Q1 = df['MonthlyIncome'].quantile(0.25)
Q3 = df['MonthlyIncome'].quantile(0.75)

# tính IQR 
IQR = Q3-Q1

upper= Q3 + 1.5 * IQR
lower = Q1 - 1.5 * IQR
```

```python
# Thực hiện loại bỏ các dòng dữ liệu ngoại lai của cột MonthlyIncome
df2 = df[~((df['MonthlyIncome'] < lower) | (df['MonthlyIncome'] > upper))]
```

```python
# Chia dữ liệu ở các cột thành 4,5,6 nhóm có số lượng phần tử bằng nhau và đếm số lượng phần tử ở mỗi nhóm, lấy ra khoảng giữ liệu của mỗi nhóm.
# Có thể chia dữ liệu theo nhiều cột khác nhau, ví dụ như theo tuổi (age) hoặc monthly income

# Chia dữ liệu theo MonthlyIncome
# 4 nhóm có số lượng phần tử bằng nhau
income = pd.qcut(df['MonthlyIncome'],4)
pd.value_counts(income)
```

```python
# 5 nhóm có số lượng phần tử bằng nhau
income = pd.qcut(df['MonthlyIncome'],5)
pd.value_counts(income)
```

```python
# 6 nhóm có số lượng phần tử bằng nhau
income = pd.qcut(df['MonthlyIncome'],6)
pd.value_counts(income)
```

- Mình nhận ra mặc dù dùng qcut thì só lượng phần tử mỗi nhóm cũng ko chính xác là bằng nhau
- Có vẻ như chia càng nhiều nhóm thì số lượng mỗi nhóm tương đồng hơn

```python
# Chia dữ liệu ở các cột age và MonthlyIncome thành 5 nhóm theo các khoảng: 0, 30, 40, 50, 80, 150 và đếm số lượng phần tử ở mỗi nhóm.
# Tạo các khoảng cho cột 'Age'
age_bins = [0, 30, 40, 50, 80, 150]

# đặt tên cho bins
age_labels = ['<30', '30-40', '40-50', '50-80', '80+']

#chia nhóm
pd.cut(df['age'], age_bins, labels=age_labels)
```
