# Thêm cột

```python
# Thêm cột số
df['rating'] = df['rating_good']*2 + df['rating_normal'] - df['rating_bad']*3
df[['rating', 'rating_good', 'rating_normal', 'rating_bad']]
```

```python
df['rating_2'] = df['rating_good']- df['rating_bad']
df[['rating_2', 'rating_good', 'rating_bad']]
```

```python
#  Thêm cột chữ
df['date'] = df['join_day'].astype('str') + "-" + df['join_month'] + "-" + df['join_year'].astype('str')
df[['date', 'join_day', 'join_month', 'join_year']]
```

# Tách cột

```python
# Tách cột chữ
#df[ten cot].str.split(kyhieu).str[index_gia_tri]
df['District'] = df['shop_location'].str.split(',').str[0]
df['City'] = df['shop_location'].str.split(',').str[1]

df[['District', 'City']]
```

```python
df['day']=df['date'].str.split('-').str[0]
df['month']=df['date'].str.split('-').str[1]
df['year']=df['date'].str.split('-').str[2]
df[['date', 'day', 'month', 'year']]
```
